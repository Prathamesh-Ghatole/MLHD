This notebook compares the performance between pandas.Series.isin() function and a custom function using Python sets and pandas apply function

```python
import pandas as pd
import lib.io_ as io
```

```python
%%time
MB_rec_gid = pd.read_parquet('warehouse/MB_tables/recording_gid.parquet')
MB_rec_gid.set_index('gid', inplace=True)
```

```python
df = io.load_path_file('warehouse/samples/random_file_paths.txt')
df
```

# Method 1: pandas.Series.isin()

```python
%%time

df[df.recording_MBID.isin(MB_rec_gid.index)]
```

# Method 2: Using sets and for loop

```python
%%time
rec_gid_set = set(MB_rec_gid.index)
```

```python
%%time
boolean_map = []
for mbid in df.recording_MBID:
    boolean_map.append(mbid in rec_gid_set)
```

# Method 3: Using sets and apply function

```python
%%time
df[df.recording_MBID.apply(lambda x: x in rec_gid_set)]
```
