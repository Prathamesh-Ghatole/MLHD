```python
import lib.postgres_conn as postgres_conn
import pandas as pd
```

```python
%%time
with postgres_conn.get_con() as conn:
    MB_recording = pd.read_sql('select gid, name from recording', con=conn)
MB_recording.head()
```

```python
%%time
with postgres_conn.get_con() as conn:
    MB_artists = pd.read_sql('select gid, name from artist', con=conn)

MB_artists.head()
```

```python
# fetch recording_gid_redirect

with postgres_conn.get_con() as conn:
    MB_redirects = pd.read_sql('select r.gid AS new, rgr.gid AS old from recording r join recording_gid_redirect rgr on rgr.new_id = r.id;', con=conn)

MB_redirects.head()
```

```python
# fetch canonical_recording table
with postgres_conn.get_con() as conn:
    MB_canonical = pd.read_sql('select recording_mbid as old, canonical_recording_mbid as new from mapping.canonical_recording_redirect', con=conn)

MB_canonical.head()
```

```python
# fetch artist_gid_redirect table

with postgres_conn.get_con() as conn:
    MB_artist_redirects = pd.read_sql('select a.gid AS new, agr.gid AS old from artist a join artist_gid_redirect agr on agr.new_id = a.id', con=conn)

# MB_artist_redirects.set_index('old', inplace=True)
MB_artist_redirects.head()
```

# Fetching names for recording-MBIDs

## Cleaning recording-MBIDs

```python
%%time
rec_replacements = MB_redirects[MB_redirects.old.isin(df2['recording-MBID'].unique())]

rec_replacements.set_index('old', inplace=True)
rec_replacements
```

```python
%%time

# Replacing rec-MBIDs with redirects in df2
# Note: actual recording-MBID column is overridden by the redirects
df2['recording-MBID'] = (df2['recording-MBID'].map(lambda x: rec_replacements.at[x, 'new'] if x in rec_replacements.index else x))
```

## Replacing non-canonical recording-MBIDs

```python
# Fetching all replacements

canonical_replacements = MB_canonical[MB_canonical.old.isin(df2['recording-MBID'].unique())]
canonical_replacements.set_index('old', inplace=True)
canonical_replacements
```

```python
# Replacing recording-MBIDs with their canonical counterparts.

df2['recording-MBID'] = df2['recording-MBID'].map(lambda id: canonical_replacements.at[id, 'new'] if id in canonical_replacements.index else id)
```

## Fetching recording_names

```python
%time

# Fetching recording name from MB_recording.
rec_names = MB_recording[MB_recording.gid.isin(df2['recording-MBID'].unique())]
rec_names
```

```python
%%time
df2_rec_name = df2.join(rec_names.set_index('gid'), on='recording-MBID')
df2_rec_name.rename({'name': 'MLHD_rec_name'}, inplace=True, axis=1)

df2_rec_name.MLHD_rec_name.isna().value_counts()
```

```python
df2_rec_name.head()
```

## Cleaning up artist_mbid with redirects.

```python
# Fetching replacements
# df2['recording-MBID'] = (df2['recording-MBID'].map(lambda x: rec_replacements.at[x, 'new'] if x in rec_replacements.index else x))

artist_replacements = MB_artist_redirects[MB_artist_redirects.old.isin(df2_rec_name['artist-MBID'].unique())]
artist_replacements.set_index('old', inplace=True)
artist_replacements
```

```python
# Fetching replacement artist MBIDs
# NOTE This process overwrites existing MBIDs.

df2_rec_name['artist-MBID'] = df2_rec_name['artist-MBID'].map(lambda id: artist_replacements.at[id, 'new'] if id in artist_replacements.index else id)
```

## Fetching artist_names for cleaned MBIDs

```python
# Fetching artist name from MB_artists

artist_names = MB_artists[MB_artists.gid.isin(df2_rec_name['artist-MBID'].unique())]
artist_names
```

```python
# Making a new table with artist names joined.

df2_rec_artist_name = df2_rec_name.join(artist_names.set_index('gid'), on='artist-MBID')
df2_rec_artist_name.rename({'name': 'MLHD_artist_name'}, inplace=True, axis=1)

df2_rec_artist_name
```

```python
df2_rec_artist_name.MLHD_artist_name.isna().value_counts()
```

# Fetching looked-up artist_credit lists respective to recording-MBIDs

```python
query_artist_list = "select recording.gid as rec_gid, array_agg(artist.gid) as artist_credit_list from recording join artist_credit ac on ac.id=artist_credit join artist_credit_name acn on acn.artist_credit=ac.id join artist on artist.id = acn.artist where recording.gid in {} group by recording.gid;".format(tuple(df2_rec_artist_name['recording-MBID'].unique()))

with postgres_conn.get_con() as conn:
    df_artist_list = pd.read_sql(query_artist_list, con=conn)

df_artist_list.artist_credit_list = df_artist_list.artist_credit_list.apply(lambda x: x.strip('{}'))
df_artist_list
```

```python
df2_artist_rec_names_artist_list = df2_rec_artist_name.join(df_artist_list.set_index('rec_gid'), on='recording-MBID')

df2_artist_rec_names_artist_list.to_csv('unk_ids/df2_artist_rec_names_artist_list.txt', index=False)
df2_artist_rec_names_artist_list
```

```python
print("% artist_name not null: {}".format(df2_artist_rec_names_artist_list.MLHD_artist_name.notna().sum() / df2_artist_rec_names_artist_list.shape[0] * 100))
print("% rec_name not null: {}".format(df2_artist_rec_names_artist_list.MLHD_rec_name.notna().sum() / df2_artist_rec_names_artist_list.shape[0] * 100))
print("% artist_credit_list not null: {}".format(df2_artist_rec_names_artist_list.artist_credit_list.notna().sum() / df2_artist_rec_names_artist_list.shape[0] * 100))
```

# Analyzing MLHD data w/ MB data

```python
%%time
# Checking if artist-MBID is in artist_credit_list

input_existance = df2_artist_rec_names_artist_list[['artist-MBID', 'artist_credit_list']].dropna()
existance_test = lambda row: (row['artist-MBID']) in (row['artist_credit_list'])

# Applying test on each artist-MBID for a boolean map of input.

existance_map = input_existance.apply(existance_test, axis=1)

print(f"% inaccurate artist-MBIDs: {(existance_map.value_counts()[0] / existance_map.value_counts().sum())*100:.2f}%")
```

```python
incorrect_artist_mbid = df2_artist_rec_names_artist_list.dropna(subset=['artist-MBID', 'artist_credit_list'])[-existance_map]
incorrect_artist_mbid
```

```python
# Exporting incorrect artist MBIDs to CSV.
incorrect_artist_mbid.to_csv('unk_ids/incorrect_artist_mbid.txt', index=False)
```
